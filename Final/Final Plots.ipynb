# Scatterplot

```python
from pandas import Timedelta
def try_timedelta(x):
    try:
        return pd.Timedelta(x)
    except:
        return None

df['clean_duration']= df['Duration'].apply(lambda x: try_timedelta(x) )
```

```python

scatter_df = df[np.logical_and(~df['clean_duration'].isna(),df['clean_duration']>pd.Timedelta(0))]
scatter_df = scatter_df[scatter_df['clean_duration'].astype(int)<10000000000000]

plt.figure(figsize=(20,10))

plt.scatter(scatter_df['clean_duration'].astype(int)/1000000000,scatter_df['Shape'].fillna("Unknown"))
plt.title("Duration of UFO Sightings by Shape in Seconds",size=20)
plt.xlabel("Duration in seconds",size=16)
plt.xticks(size=14)
plt.ylabel("Shape",size=16)
plt.yticks(size=14)
```

# Bubble Map

```python
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
bubble_map_df = df[['City','State']].groupby(['City','State'],as_index=False).size().to_frame()
bubble_map_df.reset_index(inplace=True)
```

```python
bubble_map_df = bubble_map_df[bubble_map_df[0]>2]
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="specify_your_app_name_here")
lat = []
long = []
for index,row in bubble_map_df.iterrows():
    location = geolocator.geocode(row['City']+","+row['State'], timeout=None)
    if location is None:
        lat.append(None)
        long.append(None)
    else:
        lat.append(location.latitude)
        long.append(location.longitude)
```

```python
bubble_map_df['lat']=lat
bubble_map_df['long']=long

bubble_map_df[np.logical_and(bubble_map_df['City']=="Oakland",bubble_map_df['State']=="CA")]
```

```python
#Manual adjustments for some incorrect lat/longs
bubble_map_df.loc[1411,'lat']=37.8044
bubble_map_df.loc[1411,'long']=-122.2711
bubble_map_df.loc[608,'lat']=38.2494
bubble_map_df.loc[608,'long']=-122.0400
bubble_map_df.loc[1955,'lat']=33.8358
bubble_map_df.loc[1955,'long']=-118.3406
bubble_map_df.loc[1091,'lat']=33.7701
bubble_map_df.loc[1091,'long']=-118.1937
```

```python
#bubble_map_df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv')
#bubble_map_df.head()

bubble_map_df['text'] = bubble_map_df['City']+","+bubble_map_df['State'] + '<br>UFO Sightings: ' + (bubble_map_df[0]).astype(str)
limits = [(3,5),(5,10),(10,26)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)"]#,"rgb(255,133,27)","lightgrey"]
cities = []
scale = 5000

for i in range(len(limits)):
    lim = limits[i]
    bubble_map_df_sub = bubble_map_df[np.logical_and(bubble_map_df[0]>=lim[0],bubble_map_df[0]<lim[1])]
    city = go.Scattergeo(
        locationmode = 'USA-states',
        lon = bubble_map_df_sub['long'],
        lat = bubble_map_df_sub['lat'],
        text = bubble_map_df_sub['text'],
        marker = go.scattergeo.Marker(
            size = bubble_map_df_sub[0]*10,
            color = colors[i],
            line = go.scattergeo.marker.Line(
                width=0.5, color='rgb(40,40,40)'
            ),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    cities.append(city)

layout = go.Layout(
        title = go.layout.Title(
            text = 'UFO Sightings by City'
        ),
        showlegend = True,
        geo = go.layout.Geo(
            scope = 'usa',
            projection = go.layout.geo.Projection(
                type='albers usa'
            ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        )
    )

fig = go.Figure(data=cities, layout=layout)
py.iplot(fig, filename='d3-bubble-map-ufo-sightings')
```

# Heat Map

```python
df['day_of_week'] = pd.to_datetime(df['Date / Time']).dt.dayofweek
df['morning'] = np.logical_and(pd.to_datetime(df['Date / Time']).dt.hour >= 5,pd.to_datetime(df['Date / Time']).dt.hour <12).astype(int)
df['afternoon'] = np.logical_and(pd.to_datetime(df['Date / Time']).dt.hour >= 12,pd.to_datetime(df['Date / Time']).dt.hour <19).astype(int)
df['night']=pd.to_datetime(df['Date / Time']).dt.hour.apply(lambda x: x not in list(range(5,19))).astype(int)
```

```python
heatmap_df = df[['day_of_week','morning','afternoon','night']]
```

```python
heatmap_df = heatmap_df.groupby(['day_of_week','morning','afternoon','night']).size().to_frame().reset_index()
```

```python
morning = heatmap_df[heatmap_df['morning']==1].sort_values(by="day_of_week")[0]
afternoon = heatmap_df[heatmap_df['afternoon']==1].sort_values(by="day_of_week")[0]
night = heatmap_df[heatmap_df['night']==1].sort_values(by="day_of_week")[0]
```

```python
trace = go.Heatmap(z=[morning, afternoon, night],
                   x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday','Sunday'],
                   y=['Morning', 'Afternoon', 'Night'])
data=[trace]
layout = go.Layout(
    title = go.layout.Title(
        text = 'UFO Reportings by Time of Week'
    ),
)
py.iplot(data,layout=layout, filename='labelled-heatmap')
```

# Tree Mapping

```python
import squarify

shape_of_ufo = df.groupby('Shape',as_index=False).size().sort_values(ascending=False)
plot_df = pd.DataFrame({"Shape":shape_of_ufo.index.values, "Count":shape_of_ufo.values})
plt.figure(figsize=(20,10))

SMALL_SIZE = 13
MEDIUM_SIZE = 18
BIGGER_SIZE = 23

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)   # fontsize of the figure title

cm = plt.get_cmap('Spectral')
colors = [cm((i)/20) for i in range(20)]
squarify.plot(sizes=plot_df['Count'], label=plot_df['Shape'], alpha=.8, color=colors )
plt.axis('off')
plt.title("Number of Reported UFO Shapes by Size")
plt.show() 
```
